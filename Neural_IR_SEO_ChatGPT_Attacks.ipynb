# Part 3. Dual Embedding Space Model Implementation (50 points)

In this part, we will implement the [dual embbeding space model](https://arxiv.org/pdf/1602.01137.pdf) on the genius lyrics data **lyrics_200.jl** provided in homework 1.
```python
import json
import numpy as np
from gensim.models import Word2Vec
from scipy.spatial.distance import cosine


class LyricsSearchEngine:
    def __init__(self, file_path):
        self.data = self.load_data(file_path)
        self.documents = self.preprocess_documents(self.data)
        self.model = self.train_word2vec_model(self.documents)

    @staticmethod
    def load_data(file_path):
        with open(file_path, "r") as f:
            input = []
            for line in f:
                input.append(json.loads(line))
        return input

    @staticmethod
    def preprocess_documents(input):
        new_doc = []
        for line in input:
            words = line["lyrics"].lower().split()
            new_doc.append(words)
        return new_doc

    @staticmethod
    def train_word2vec_model(documents):
        model = Word2Vec(documents, vector_size=200, window=5, min_count=1, workers=5, epochs=20, seed=1)
        return model

    def get_embedding(self, words):
        word_embeddings = []
        for word in words:
            if word in self.model.wv:
                word_embeddings.append(self.model.wv[word])
        if word_embeddings:
            return np.mean(word_embeddings, axis=0)
        else:
            return np.zeros(self.model.vector_size)


    def get_top_results(self, query, k):

        query_tokens = query.lower().split()
        query_embedding = self.get_embedding(query_tokens)

        doc_embeddings = []
        for doc in self.documents:
          doc_embedding = self.get_embedding(doc)
          doc_embeddings.append(doc_embedding)

        scores = []
        for doc_embedding in doc_embeddings:
            score = 1 - cosine(query_embedding, doc_embedding)
            scores.append(score)

        top_scored_indices = np.argsort(scores)[-k:][::-1]
        top_results = []
        for id in top_scored_indices:
            song = self.data[id]["song"]
            score = scores[id]
            lyrics = self.data[id]["lyrics"].replace('\n', ' ')
            top_results.append((song, score, lyrics))

        return top_results




    # Initialize the search engine
if __name__ == '__main__':

    search_engine = LyricsSearchEngine('/content/lyrics_200.jl')
    # Retrieve top-k results for each query
    querie_list = ["time", "never know", "make no sense"]
    for query in querie_list:
        print(f"Query: '{query}'")
        print("-------------------------------------------------------------------")
        top_results = search_engine.get_top_results(query, k=5)
        print("Rank | Score | Document Id----------------------------------|------------Document")
        for rank, result in enumerate(top_results, start=1):
            song, score, lyrics = result
            print(f"{rank}) {score:.5} |{song} | {lyrics}")
        print()
    
```

## Discussion (10 points)

Compare the retrieval performance of DESM against BM25 you got in HW1 Part C. Briefly discuss the differences you see between DESM and BM25.

DualEmbeddingSpaceModel (DESM) and BM25 are both information retrieval models used to retrieve relevant documents based on a user's query. However, there are some differences between them.

BM25 is a bag-of-words model that ranks documents based on the relevance score between the query terms and the terms in the document. It takes into account factors like term frequency, inverse document frequency, and document length normalization. BM25 is a widely used information retrieval model and is known to perform well for short queries and relatively small document collections.

On the other hand, DESM is a neural network-based model that represents both the query and the document in a shared embedding space. It learns the relationship between the query and the document in this embedding space, which allows it to capture more complex relationships between the query and the document. DESM has been shown to perform well for long queries and large document collections.

In terms of retrieval performance, both BM25 and DESM have been shown to perform well in different scenarios. For example, in some studies, BM25 has been shown to outperform DESM for short queries and relatively small document collections, while DESM has been shown to outperform BM25 for long queries and large document collections.

Overall, the choice of model depends on the specific needs of the application and the characteristics of the dataset. If the queries are short and the document collection is small, BM25 may be a better choice. If the queries are long and the document collection is large, DESM may be a better choice.
